```python
#make the plot summarising the results now..
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)

theta = [np.pi/3]
radii = [0.5]
width = np.pi/3
bars = ax.bar(theta, radii, color='lime',width = width, alpha=0.4)

patterns = ('.',)
for bar, pattern in zip(bars, patterns):
    bar.set_hatch(pattern)

theta = [2*np.pi/3]
radii = [0.5]
width = np.pi/3
bars = ax.bar(theta, radii, color='gold',width = width, alpha=0.4)

patterns = ('.',)
for bar, pattern in zip(bars, patterns):
    bar.set_hatch(pattern)
    

ax.set_yticks([.1,.2,.3,.4,.5])
ax.set_ylim([0, 0.5])
ax.set_xticks(np.pi/180. * np.linspace(180,  -180, 12, endpoint=False))
ax.set_thetamin(30)
ax.set_thetamax(150)
ax.set_xticklabels(['-90$^\circ$','-60$^\circ$','left','0$^\circ$','right','60$^\circ$',
                    '90$^\circ$','-90$^\circ$','-60$^\circ$','left','0$^\circ$','right'])
ax.set_yticklabels([])


plt.savefig(fig_dir + "stimulus_angles.png", bbox_inches = 'tight', pad_inches = 0.1, dpi=240, transparent=True)
```
